```python
import xml.etree.ElementTree as ET
import myutils
import myessentia
import essentia
import essentia.standard
import librosa
import os.path
import madmom
import numpy as np        
    
def getOnsetsFromFile(filename):
    root = ET.parse(filename).getroot()

    transcriptions = root[1]

    onsets = {
        "kicks" : [],
        "snares" : [],
        "hats" : [],
    }

    for event in transcriptions:
        if event[3].text == "KD":
            onsets["kicks"].append(float(event[1].text))
        if event[3].text == "SD":
            onsets["snares"].append(float(event[1].text))
        if event[3].text == "HH":
            onsets["hats"].append(float(event[1].text))
            
    return onsets

def segmentFile(inputWavFile, outputPath, onsetTimes):
    audio, sampleRate = librosa.load(inputWavFile, sr=44100)

    startTimes, endTimes = myessentia.getSlicerTimesFromOnsets(audio, onsetTimes)
    
    slicer = essentia.standard.Slicer()
    
    pass

    

def processFiles(inputPath, outputPath):
    xmlFiles = myutils.getListOfFiles(inputPath + "/annotation_xml", "*.xml")
    
    for xmlFile in xmlFiles:
        onsets = getOnsetsFromFile(xmlFile)
        
        justTheFilename = os.path.basename(xmlFile).split("#")[0]
        
        kickWavFile = "%s/audio/%s%s" % (inputPath, justTheFilename, "#KD#train.wav")                    
        segmentFile(kickWavFile, outputPath + "/kicks", onsets["kicks"])
        
        snareWavFile = "%s/audio/%s%s" % (inputPath, justTheFilename, "#SD#train.wav")                    
        segmentFile(snareWavFile, outputPath + "/snares", onsets["snares"])
        
        hatWavFile = "%s/audio/%s%s" % (inputPath, justTheFilename, "#HH#train.wav")                    
        segmentFile(hatWavFile, outputPath + "/hats", onsets["hats"])
            
processFiles("/Users/carthach/tmp/drum_timbre/idmt/IDMT-SMT-DRUMS-V2/", "/Users/carthach/tmp/drum_timbre/dataset")
```
