```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

SMALL_SIZE = 11
MEDIUM_SIZE = 13
BIGGER_SIZE = 15

# plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
# plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
# plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
# plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
# plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
# plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
# plt.rc('figure', titlesize=MEDIUM_SIZE)  # fontsize of the figure title

# plt.figure(figsize=(10.1, 9.1))
# plt.figure(figsize=(5.16, 2), dpi=150)

def makePlot(bestFilename, avgFilename, key, title, xLabel=False, yLabel=False):
    plt.subplot(key)
    
    plt.tight_layout()
    
    best = np.loadtxt(bestFilename)
    avg = np.loadtxt(avgFilename)
    
    plt.title(title)
    
    plt.plot(avg, label="Average Fitness")
    plt.plot(best, label="Best Fitness", marker='*', linestyle='--', color='r', markersize=7, alpha=0.5)
    
    if xLabel:
        plt.xlabel("Generations")
    if yLabel:
        plt.ylabel("Fitness")
        
    plt.legend()

makePlot("hamming_linear_best.txt", "hamming_linear_avg.txt", 221, "Hamming-Linear", xLabel=False, yLabel=True)
makePlot("hamming_parallel_best.txt", "hamming_parallel_avg.txt", 222, "Hamming-Parallel", xLabel=False, yLabel=False)
makePlot("swap_linear_best.txt", "swap_linear_avg.txt", 223, "Swap-Linear", xLabel=True, yLabel=True)
makePlot("swap_parallel_best.txt", "swap_parallel_avg.txt", 224, "Swap-Parallel", xLabel=True, yLabel=False)

plt.savefig("best-avg_fitness.pdf", bbox_inches='tight', dpi=1200)
plt.show()
```
