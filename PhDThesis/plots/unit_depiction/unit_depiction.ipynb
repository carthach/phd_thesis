```python
import matplotlib.pyplot as plt
import numpy as np
import essentia.standard
import librosa.display
import librosa

```

```python
#Load Audio
loader = essentia.standard.MonoLoader(filename="amen_break.mp3")
audio = loader()

audio = audio[0:44100]

onsetRateExtractor = essentia.standard.OnsetRate()

onsetTimes, onsetRate = onsetRateExtractor(audio)

plt.plot(audio)

onsetTimes *= 44100

for onsetTime in onsetTimes:
    plt.axvline(x=onsetTime, color='r', linestyle='--')
    plt.figure
    plt.margins(0)
    
    plt.xticks([])
    plt.yticks([])    
    
plt.savefig("combined.png", bbox_inches='tight')
    
onsetTimes = np.append(onsetTimes, [len(audio)])
onsetTimes = onsetTimes.astype(int)

print(onsetTimes)

plt.figure()
plt.subplots_adjust(wspace = 0.75)

for i in range(4):
    xStart = onsetTimes[i]
    xEnd = onsetTimes[i+1]
            
    subPlotIndex = '14%d' % i
    plt.subplot(subPlotIndex)
    plt.plot(audio[xStart: xEnd])
    plt.margins(0)
    
    plt.xticks([])
    plt.yticks([])    
    
    filename = "unit_%d.png" % i
    
    plt.savefig(filename, bbox_inches='tight')
    
    
plt.savefig('decomposition.png', bbox_inches='tight')
    
plt.show()

# step = 44100

# start = step * 2

# j = 0
# for i in range(4):
#     plt.figure()
#     plt.xticks([])
#     plt.yticks([])
#     plt.margins(0)
#     plt.plot(audio[start:start+step], color='black')
#     filename = "target_%d.png" % j
#     plt.axis('tight')
#     plt.savefig(filename, bbox_inches='tight')
#     j += 1
    
#     start+=step
    
# # # plt.plot(audio[i:i+step], color='red')
# # # plt.savefig('corpus_%d' % j)
# # # j += 1
```
